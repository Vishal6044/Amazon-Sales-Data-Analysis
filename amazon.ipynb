```python
#Category Wise Sales

category_sales=df.groupby("Category")["Total Sales"].sum()
print(category_sales)
```

```python
#Visualization

plt.figure(figsize=(8,5))
category_sales.plot(kind="bar",color="skyblue")
plt.title("Category Wise Sales")
plt.xlabel("Category")
plt.ylabel("Total Sales")
plt.show()
```

```python
#Product Wise Sales

producat_sales=df.groupby("Product")["Total Sales"].sum()
print(producat_sales.sort_values(ascending=False))
```

```python
#Top 10 Products

producat_sales.sort_values(ascending=False).head(10).plot(kind="bar")
plt.show()
```

```python
#Payment Method Analysis

print(df["Payment Method"].value_counts())
```

```python
sns.countplot(x="Payment Method",data=df)
plt.show()
```

```python
#Order Status

print(df["Status"].value_counts())
```

```python
sns.countplot(x="Status",data=df)
plt.show()
```

```python
#City Wise Order
print(df["Customer Location"].value_counts())
```

```python
#Top 10
df["Customer Location"].value_counts().head(10).plot(kind="bar")
plt.show()
```

```python
#Correlation
df.corr(numeric_only=True)
```

```python
#Headmap

plt.Figure(figsize=(8,5))
sns.heatmap(df.corr(numeric_only=True)),
annot=True
camp="Coolwarm"
plt.show()
```

```python
#Sales vs Quantity

plt.figure(figsize=(8,5))
sns.scatterplot(x="Quantity",
                y="Total Sales",
                data=df)
plt.show()
```

```python
#Distribuation of Sales

plt.Figure(figsize=(8,5))
sns.histplot(df["Total Sales"],bins=20)
plt.show()
```

```python
# Box Plot (Outliners)

plt.figure(figsize=(8,5))
sns.boxplot(y="Total Sales",data=df)
plt.show()
```
